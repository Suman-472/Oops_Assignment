```python
# 7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

## Method Resolution Order (MRO): MRO in Python is the order in which methods or attributes are searched for when an object is accessed.
                                # It's crucial in understanding how inheritance works in Python, especially when dealing with multiple inheritance.
                                # Python uses the C3 linearization algorithm to determine the MRO for a class.This algorithm ensures that:
                                    #1. Parent classes are visited before their subclasses.
                                    #2. A class appears only once in the MRO.
                                    #3. If a class inherits from multiple parents, the MRO is constructed such that the leftmost parent's MRO is considered first.
          # Example:
                   class A:
                       pass

                   class B(A):
                       pass

                   class C(A):
                       pass

                   class D(B, C):
                       pass
                   print(D.__mro__)   # Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
                   # In this example, the MRO for class D is:
                                                           # 1. D itself
                                                           # 2. B (leftmost parent)
                                                           # 3. C (rightmost parent)
                                                           # 4. object (base class for all Python objects)

      # To retrieve the Method Resolution Order (MRO) of a class programmatically in Python, you can access the __mro__ attribute.
         # This attribute returns a tuple containing the classes in the MRO, starting with the class itself and ending with the base class object.
                # Example:
                         class A:
                             pass

                         class B(A):
                             pass

                         class C(A):
                             pass

                         class D(B, C):
                             pass

                         # Get the MRO of class D
                         mro = D.__mro__

                         print(mro)  # Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
                         # As we can see, the MRO for class D includes D itself, its parent classes B and C, the common ancestor A, and the base class object.
```

```python
# 9. Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.\

## calculates and prints the areas of different shape objects:
    class Shape:
        def area(self):
            raise NotImplementedError("Subclasses must implement area()")

    class Circle(Shape):
        def __init__(self, radius):
          self.radius = radius

        def area(self):
            return 3.14159 * self.radius ** 2

    class Rectangle(Shape):
        def __init__(self, length, width):
           self.length = length
           self.width = width

        def area(self):
            return self.length * self.width

    class Triangle(Shape):
        def __init__(self, base, height):
           self.base = base
           self.height = height

        def area(self):
            return 0.5 * self.base * self.height

        def calculate_and_print_area(shape):
            print(f"Area of {shape.__class__.__name__}: {shape.area():.2f}")

# Create instances of different shapes
circle = Circle(4)
rectangle = Rectangle(9, 6)
triangle = Triangle(4, 6)

# Calculate and print their areas using the same function
calculate_and_print_area(circle)       # Output: Area of Circle: 50.27
calculate_and_print_area(rectangle)    # Output: Area of Rectangle: 54.00
calculate_and_print_area(triangle)     # Output: Area of Triangle: 12.00
# Explanation:
     #1. Base Class Shape:
                  # Defines a common interface with the area() method, which is declared as abstract using NotImplementedError.
                  # This ensures that subclasses must implement their own area() method.
     #2. Derived Classes:
                  # Circle, Rectangle, and Triangle inherit from the Shape class.
                  # Each class implements its own area() method to calculate the area specific to its shape.
     #3. calculate_and_print_area() Function:
                  # Takes a shape object as an argument.
                  # Calls the area() method on the shape object to calculate its area.
                  # Prints the area along with the shape's name.
     #4. Creating and Using Shapes:
                  # Instances of Circle, Rectangle, and Triangle are created with their respective parameters.
                  # The calculate_and_print_area() function is called for each shape, demonstrating polymorphism.
```

```python
# 10. Implement encapsulation in a 'BankAccount' class with private attributes for 'balance' and 'account_number'. Include methods for deposit, withdrawal, and balance inquiry.

## Here’s an example of how you can implement encapsulation in a BankAccount class in Python.
     # This class will have private attributes for balance and account_number, and methods for deposit, withdrawal, and balance inquiry:
        class BankAccount:
            def __init__(self, account_number, initial_balance=0):
                self.__account_number = account_number
                self.__balance = initial_balance

            def get_account_number(self):
                return self.__account_number

            def deposit(self, amount):
                if amount > 0:
                   self.__balance += amount
                   print(f"Deposited {amount}. New balance is {self.__balance}.")
               else:
                   print("Deposit amount must be positive.")

            def withdraw(self, amount):
                if 0 < amount <= self.__balance:
                   self.__balance -= amount
                   print(f"Withdrew {amount}. New balance is {self.__balance}.")
               else:
                   print("Insufficient balance")

            def get_balance(self):
                return self.__balance
        account = BankAccount("258997442", 10000)
        print("Account Number:", account.get_account_number())
        account.deposit(5000)
        account.withdraw(8600)
        print("Current Balance:", account.get_balance())
        # Output:
               Account Number: 258997442
               Deposited 5000. New balance is 15000.
               Withdrew 8600. New balance is 6400.
               Current Balance: 6400
```

```python
# 13.  Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

## Diamond problem:
                  # The diamond problem is a specific issue that can arise in programming languages that support multiple inheritance, including Python.
                  # It occurs when a class inherits from two or more classes that have a common base class.
                  # This can lead to ambiguity in method resolution, causing conflicts and making it unclear which version of a method should be used.

    # How Python Resolves the Diamond Problem:
                  # To mitigate the diamond problem, Python uses a method resolution order (MRO) algorithm called C3 linearization.
                  # The MRO determines the order in which the base classes are searched for a method or attribute.
                  # It follows a specific set of rules to ensure a consistent and unambiguous order.
    # Here’s an example that demonstrates the diamond problem and how Python resolves it:
                 # Class A:
                          class A:
                              def method(self):
                                  print("Method in A")
                          class B(A):
                              def method(self):
                                  print("Method in B")
                          class C(A):
                              def method(self):
                                  print("Method in C")
                          class D(B, C):
                              pass
                          d = D()
                          d.method()     # Output: Method in B
          # Method in B:
                      # In this example, we have four classes: A, B, C, and D. Both B and C inherit from A, and D inherits from both B and c. All classes define a method called method().
                      # When d.method() is called, Python follows the MRO to determine the order in which the base classes are searched for the method.

                      # In this case, the MRO is D -> B -> C -> A. So, Python looks for the method() in class D first, then in B, then in C, and finally in A.
                      # As a result, the output is “Method in B”. This is because B is the first class in the MRO that defines the method(), so its implementation is used.

                      # Python’s MRO algorithm effectively resolves the diamond problem by providing a well-defined order for method resolution.
                      # However, it’s important to be aware of potential conflicts and understand the MRO to design your class hierarchy and override methods appropriately.
```
